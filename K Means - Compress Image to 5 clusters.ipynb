## INTRODUCTION

In this assignment students have to compress racoon grey scale image into5 clusters. In the end, visualize both raw and compressed image and look
for quality difference.

The raw image is available in spicy.misc package with the name face.

```python
# Load all the necessary libraries
import numpy as np
from sklearn.cluster import KMeans
import scipy.misc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Visualize the gray scale image

f = scipy.misc.face(gray=True)
plt.figure(figsize=(10, 3.6))
plt.imshow(f, cmap=plt.cm.gray)
plt.show()
```

```python
# Retrieve the number of rows and columns for the picture

rows = f.shape[0]
cols = f.shape[1]
print("The number of rows is : ", rows)
print("The number of cols is : ", cols)
```

```python
# Compressing the gray scale image into 5 clusters 
image = f.reshape(rows*cols,1)
kmeans = KMeans(n_clusters = 5)
kmeans.fit(image)
```

```python
clusters = np.asarray(kmeans.cluster_centers_) 
labels = np.asarray(kmeans.labels_)  
labels = labels.reshape(rows,cols); 
```

```python
plt.imsave('compressed_racoon.png',labels);
```

```python
# Visualize the compressed image

image = plt.imread('compressed_racoon.png')
plt.figure(figsize=(10, 3.6))
plt.imshow(image)
plt.show()
```
