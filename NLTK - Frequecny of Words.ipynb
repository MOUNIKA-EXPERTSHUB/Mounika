## INTRODUCTION

In this assignment students have to find the frequency of words in awebpage. User can use urllib and BeautifulSoup to extract text from webpage.

Webpage : https://www.php.net/

```python
# Load all the necessary libraries
from bs4 import BeautifulSoup
import urllib.request
import nltk
```

### Read the webpage

```python
response = urllib.request.urlopen('http://php.net/')
html = response.read()
```

### Use Beautiful Soup to strop of all the HTML tags from the file

```python
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
```

```python
# Display the Text that is being captured.
text
```

### Convert that text into tokens by splitting the text

Here we will try to tokenize using split and then using tokenize function in nltk package.

```python
# Using the split fucntion
token = [tkn for tkn in text.split()]
```

```python
token
```

### Tokenize using NLTK

```python
from nltk.tokenize import word_tokenize
```

```python
nltk_token = word_tokenize(text); print (nltk_token)
```

#### Remove Stop Words Using NLTK

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
# Remove all the stop words from the dataset
for i in token:
    if i in stopwords.words('english'):
        token.remove(i)
```

```python
# Remove all the stop words from the nltk dataset
for i in nltk_token:
    if i in stopwords.words('english'):
        nltk_token.remove(i)
```

#### Compute the frequency using nltk

```python
# Dataset created using split function
freq = nltk.FreqDist(token) 
```

```python
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```

```python
freq = nltk.FreqDist(nltk_token)
```

```python
for key,val in freq.items(): 
    print (str(key) + ':' + str(val))
```
