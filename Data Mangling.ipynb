```python
#Load datasets from csv

users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv') 

#Print first 5 rows for each of the dataset
print(users.head())
print(sessions.head())
print(products.head())
print(transactions.head())
```

```python
#Question 12: Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)

#Left Join transactions and users dataframe on 'UserID'
results = pd.merge(transactions, users, on='UserID', how='left')

#Print the resulting dataframe
results
```

```python
#Question 13: Which transactions have a UserID not in users?

#Get all the UserID values from users dataframe
userIds = users.loc[:,"UserID"].values

#Fitler only those records from transactions dataframe where the UserID matches with users dataframe
transactions[transactions.UserID.isin(userIds)]
```

```python
#Question 14: Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)

#Inner Join on transactions and users dataframe on 'UserID'
results = pd.merge(transactions, users, on='UserID', how='inner')

#Print the resulting dataframe
results
```

```python
#Question 15: Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)

#Full outer Join on transactions and users dataframe on 'UserID'
results = pd.merge(transactions, users, on='UserID', how='outer')

#Print the resulting dataframe
results
```

```python
#Question 16: Determine which sessions occurred on the same day each user registered

#Inner Join on users and sessions dataframe on 'UserID'
results = pd.merge(users, sessions, on='UserID', how='inner')

#Filter the results where registration date ('Registered') is equal to session date ('SessionDate')
results[results['Registered'] == results['SessionDate']]
```

```python
#Question 17: Build a dataset with every possible (UserID, ProductID) pair (cross join)

#Assign a temporary key to both the dataframes and merge based on that key. After merging, drop the key column
results = users.assign(key=1).merge(products.assign(key=1), on='key').drop('key', 1)

#Show only 'UserID' and 'ProductID' combinations
results[['UserID', 'ProductID']]
```

```python
#Question 18: Determine how much quantity of each product was purchased by each user

#Assign a temporary key to both the dataframes (users, products) and merge based on that key. After merging, drop the key column
results = users.assign(key=1).merge(products.assign(key=1), on='key').drop('key', 1)

#Merge the results dataframe with transactions dataframe on 'UserID' and 'ProductID'
user_products = pd.merge(results, transactions,  how='left', left_on=['UserID','ProductID'], right_on= ['UserID','ProductID'])

#Do group by on 'UserID' and 'ProductID' and get the sum of 'Quantity'. Display 3 columns as 'UserID', 'ProductID', 'Quantity'
user_products.groupby(['UserID', 'ProductID']).sum().reset_index()[['UserID','ProductID','Quantity']]
```

```python
#Question 19: For each user, get each possible pair of pair transactions (TransactionID1,TransacationID2)

#Assign a temporary key to transactions dataframe and merge based on that key. After merging, drop the 2 columns - key, UserID_y
results = transactions.assign(key=1).merge(transactions.assign(key=1), on='key').drop(['key', 'UserID_y'], 1)

#Rename the column 'UserID_x' to 'UserID' and update the existing dataframe with new column name using inplace = True
results.rename(columns={'UserID_x':'UserID'}, inplace = True)

#For each user, display each possible pair of pair transactions
results
```

```python
#Question 20: Join each user to his/her first occuring transaction in the transactions table

#Perform groupby 'UserID' on transactions dataframe and get the record with first transactionDate. Then left Join users on 'UserID'
results = pd.merge(users, transactions.groupby('UserID').first().reset_index(), on='UserID', how='left')

#Print the resulting dataframe showing first occurring transaction for each user
results
```

```python
#Question 21: Test to see if we can drop columns
#Get the list of all the columns in the results dataframe
my_columns = list(results.columns) 
my_columns
```

```python
#Check to see the columns that have empty values within certain threshold
list(results.dropna(thresh=int(results.shape[0] * .9), axis=1).columns)
```

```python
#Check to see what columns have empty values
missing_info = list(results.columns[results.isnull().any()])
missing_info
```

```python

#Check to see the count of empty values in the columns with missing info
for col in missing_info:
    num_missing = results[results[col].isnull() == True].shape[0] 
    print('number missing for column {}: {}'.format(col, num_missing)) #count of missing data
```

```python
#Check to see the percent of empty values in the columns with missing info
for col in missing_info:
    percent_missing = results[results[col].isnull() == True].shape[0] / results.shape[0] 
    print('percent missing for column {}: {}'.format(col, percent_missing))
```
